```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#Import data 
CAC40 = pd.read_csv("^cac_d.csv")
Bovespa = pd.read_csv("^bvp_d.csv")
DAX = pd.read_csv("^dax_d.csv")
HSI = pd.read_csv("^hsi_d.csv")
NASDAQ = pd.read_csv("^ndq_d.csv")
Nikkei = pd.read_csv("^nkx_d.csv")
Aord = pd.read_csv("^aor_d.csv")
Shanghai = pd.read_csv("^shc_d.csv")
SP500 = pd.read_csv("^spx_d.csv")
Invesco = pd.read_csv("invesco_us_d.csv")
```

```python
Invesco.head()
```

```python
#We generate an empty DataFrame with index as index of Invesco.
indicepanel = pd.DataFrame(index = Invesco.index) 

indicepanel['Invesco'] = Invesco['Open'].shift(-1) - Invesco['Open']
indicepanel['Invesco_1'] = indicepanel['Invesco'].shift(1)

indicepanel['NASDAQ'] = NASDAQ['Open'].shift(-1) - NASDAQ['Open']
indicepanel['CAC40'] = CAC40['Open'].shift(-1) - CAC40['Open']
indicepanel['Bovespa'] = Bovespa['Open'].shift(-1) - Bovespa['Open']
indicepanel['SP500'] = SP500['Open'].shift(-1) - SP500['Open']
indicepanel['DAX'] = DAX['Open'].shift(-1) - DAX['Open']

# According to differences between time zones we have to use different formula. Asian's and Australian markets data are available when US market is oppening.
indicepanel['Nikkei']=Nikkei['Close'] - Nikkei['Open']
indicepanel['HSI'] = HSI['Close'] - HSI['Open']
indicepanel['Aord'] = Aord['Close'] - Aord['Open']
indicepanel['Shanghai'] = Shanghai['Close'] - Shanghai['Open']


#we are going to keep column "Open" of Invesco to use it for evaluation.

indicepanel['Price'] = Invesco['Open']
 
```

```python
indicepanel = indicepanel.fillna(method = 'ffill')
indicepanel = indicepanel.dropna()
# we already moved values to eliminate empty cells
```

```python
#Split the data in two parts. We have 1007 days of data. 
#The resonable split will be by 500 for Train and 500 for Test
Train = indicepanel.iloc[-1000: -500, :] # to build the model
Test = indicepanel.iloc[-500:, :] # to test the model
print(Train.shape, Test.shape)
```

```python
from pandas.plotting import scatter_matrix
sm = scatter_matrix(Train, figsize=(15, 15), color = 'navy')
```

```python
# From the scatter matrix we can see that Invesco has a linear pattern with SP500 and with NASDAQ.
# The rest is non-linear pattern.
```

```python
corr_array = Train.iloc[:, :-1].corr()['Invesco']
print(corr_array)
# We can see that NASDAQ and SP500 have strong positive correlation with Invesco. 
#It means that if those two predictors increase then more likely Invesco as well increase.
```

```python
# For mor informations we can use a following formula :
y = 'Invesco~Invesco_1+NASDAQ+CAC40+Bovespa+SP500+DAX+Nikkei+HSI+Aord+Shanghai'
lm = smf.ols(formula = y, data = Train).fit()
lm.summary()
```

```python
# Probability of F-statistic is very small, almost zero. If P value < 0.05 then we can reject hipothesis H_0. 
# Hipothesis H_0 means that all ours predictors in model are useless.
# From the column with P value for individual predictors we can see that only NASDAQ, CAC40, Bovespa and DAX are significant.
```

```python
# Make prediction

Train['PredictedY'] = lm.predict(Train)
Test['PredictedY'] = lm.predict(Test)

```

```python
# Scatter plot between daily and predict daily change
plt.scatter(Train['Invesco'],Train['PredictedY'], color = 'indigo')
plt.axhline(y=0, color='red')
```

```python
# It has strong positive correlation
```

```python
def adjustedMetric(data, model, model_k, yname):
    data['yhat'] = model.predict(data)
    SST = ((data[yname] - data[yname].mean())**2).sum()
    SSR = ((data['yhat'] - data[yname].mean())**2).sum()
    SSE = ((data[yname] - data['yhat'])**2).sum()
    r2 = SSR/SST
    adjustR2 = 1 - (1 - r2)*(data.shape[0] - 1)/(data.shape[0] - model_k-1)
    RMSE = (SSE/(data.shape[0] - model_k-1))**0.5
    return adjustR2, RMSE
    
    
```

```python
def assessTable(test, train, model, model_k, yname):
    r2test, RMSEtest = adjustedMetric(test, model, model_k, yname)
    r2train, RMSEtrain = adjustedMetric(train, model, model_k, yname)
    assessment = pd.DataFrame(index=['R2', 'RMSE'], columns=['Train', 'Test'])
    assessment['Train'] = [r2train, RMSEtrain]
    assessment['Test'] = [r2test, RMSEtest]
    return assessment
```

```python
assessTable(Test, Train, lm, 10, 'Invesco')
```

```python
Train['Order'] = [1 if sig> 0 else -1 for sig in Train['PredictedY']]
Train['Profit'] = Train['Invesco'] * Train['Order']

Train['Wealth'] = Train['Profit'].cumsum()
print('Profit in 1000 days in Train is:', Train['Profit'].sum())
```

```python
plt.figure(figsize=(10, 10))
plt.title('Performance of Strategy in Train')
plt.plot(Train['Wealth'].values, color='indigo', label='Signal based strategy')
plt.plot(Train['Invesco'].cumsum().values, color='silver', label='Buy and Hold strategy')
plt.axhline(y=0, color='red')
plt.legend()
plt.show()
```

```python
Test['Order'] = [1 if sig>0 else -1 for sig in Test['PredictedY']]
Test['Profit'] = Test['Invesco'] * Test['Order']

Test['Wealth'] = Test['Profit'].cumsum()
print('Profit in 1000 days in Test is :', Test['Profit'].sum())
```

```python
plt.figure(figsize=(10, 10))
plt.title('Performance of Strategy in Test')
plt.plot(Test['Wealth'].values, color='indigo', label='Signal based strategy')
plt.plot(Test['Invesco'].cumsum().values, color='silver', label='Buy and Hold strategy')
plt.axhline(y=0, color='red')
plt.legend()
plt.show()
```

```python
Train['Wealth'] = Train['Wealth'] + Train.loc[Train.index[0], 'Price']
Test['Wealth'] = Test['Wealth'] + Test.loc[Test.index[0], 'Price']
```

```python
Train['Return'] = np.log(Train['Wealth']) - np.log(Train['Wealth'].shift(1))
dailyr = Train['Return'].dropna()

print('Daily Sharpe Ratio is ', dailyr.mean()/dailyr.std(ddof=1))
print('Yearly Sharpe Ratio is ', (252**0.5)*dailyr.mean()/dailyr.std(ddof=1))
```

```python
Test['Return'] = np.log(Test['Wealth']) - np.log(Test['Wealth'].shift(1))
dailyr = Test['Return'].dropna()

print('Daily Sharpe Ratio is ', dailyr.mean()/dailyr.std(ddof=1))
print('Yearly Sharpe Ratio is ', (252**0.5)*dailyr.mean()/dailyr.std(ddof=1))
```

```python
Train['Peak'] = Train['Wealth'].cummax()
Train['Drawdown'] = (Train['Peak'] - Train['Wealth'])/Train['Peak']
print('Maximum Drawdown in Train is ', Train['Drawdown'].max())
```

```python
Test['Peak'] = Test['Wealth'].cummax()
Test['Drawdown'] = (Test['Peak'] - Test['Wealth'])/Test['Peak']
print('Maximum Drawdown in Test is ', Test['Drawdown'].max())
```
